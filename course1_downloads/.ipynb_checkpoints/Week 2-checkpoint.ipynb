```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)

names_ids = df.index.str.split('\s\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name (new index) 
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

df = df.drop('Totals')
df.head(10)
```

```python
def answer_one():
    return df[df['Gold'] == df['Gold'].max()].index[0]
answer_one()
```

```python
def answer_three():
    max_val = 0
    df_tmp = df[(df['Gold'] > 1) & (df['Gold.1'] > 1)]
    
    df_tmp['GOLDDIFF'] = (df_tmp['Gold'] - df_tmp['Gold.1']) / (df_tmp['Gold'] + df_tmp['Gold.1'])
    
    return df_tmp[(df_tmp['GOLDDIFF'] == df_tmp['GOLDDIFF'].max())].index[0]

answer_three()
```

```python
census_df = pd.read_csv('census.csv')
census_df.head()
```

```python
def answer_five():
    columns_to_keep = ['STNAME', 'CTYNAME']
    census_df_fun = census_df[(census_df['SUMLEV'] == 50)][columns_to_keep]
    summary = census_df_fun.groupby(['STNAME']).count()
    
    return summary[summary['CTYNAME'] == summary['CTYNAME'].max()].index[0]

answer_five()
```

```python
census_df.groupby(['STNAME']).count()
```

```python
def answer_six():
    columns_to_keep = ['STNAME', 'CTYNAME', 'CENSUS2010POP']
    census_df_fun = census_df[columns_to_keep][(census_df['SUMLEV'] == 50)]
    states = census_df_fun['STNAME'].unique()
    pops = []
    
    for state in states: 
        df_fun = census_df_fun[census_df_fun['STNAME'] == state].nlargest(3, 'CENSUS2010POP')
        pops.append([state, df_fun['CENSUS2010POP'].sum()])
        
    states = sorted(pops, key=lambda x: x[1], reverse=True)[:3]
    return [i[0] for i in states]

answer_six()
```

```python
columns_to_keep = ['CTYNAME', 'POPESTIMATE2010', 
                   'POPESTIMATE2011', 
                   'POPESTIMATE2012', 
                   'POPESTIMATE2013',
                   'POPESTIMATE2014', 
                   'POPESTIMATE2015']
census = census_df[columns_to_keep][(census_df['SUMLEV'] == 50)]

census['ABSCHANGE'] = census[columns_to_keep].max(axis=1) - census[columns_to_keep].min(axis=1) 

census = census[(census['ABSCHANGE'] == census['ABSCHANGE'].max())].set_index('CTYNAME')

census.index[0]
```
